### Loading

- Load images from .dcm files

```python
def load_mask(file):
    """
    Load masks from .gz files

    :param file: input filename
    :return: array of 2D masks (z, x, y)
    """
    mask = nib.load(file)
    mask = mask.get_fdata().transpose(2, 0, 1)
    mask = np.rot90(mask, axes=(1,2))
    mask = mask.astype('int')
    
    return mask


def load_dicom(directory):
    """
    Load images from .dcm files

    :param directory: directory with .dcm files
    :return: array of 2D images (z, x, y)
    """

    reader = sitk.ImageSeriesReader()
    dicom_names = reader.GetGDCMSeriesFileNames(directory)
    reader.SetFileNames(dicom_names)
    image_itk = reader.Execute()
    image_zyx = sitk.GetArrayFromImage(image_itk).astype(np.int16)
    image_zyx = image_zyx.astype('int')
    
    return image_zyx 


def load_config(config_name):
    """
    Load configuration file
    :param config_name:
    :return: configuration data
    """
    with open(config_name) as file:
        config = yaml.safe_load(file)

    return config
```

### Image analysis

- Get directories with images

```python
images_folders = [str(name) for name in 
                      list(Path(os.path.join('../'+config['dataset']['dir_images'])).rglob("**")) 
                         if os.listdir(name)[0].endswith('.dcm')]
```

- Load and check amount and shape of images

```python
images = {}
total = 0 

for i, d in enumerate(sorted(images_folders)):
    try:
        sub_folder = re.findall(r'LUNG[0-9]*-[0-9]*', d)[0]
        images[sub_folder] = load_dicom(d)
        print(f'{images[sub_folder].shape[0]} files\tin {sub_folder} directory with shape {images[sub_folder].shape[1]}x{images[sub_folder].shape[2]}')
        total += images[sub_folder].shape[0]
    except:
        print(f'WARNING: No files with ".dcm" extention in directory {d}')
        
print(f'\nTotal amount of images: {total}')
```

```python
np.random.seed(12345)

fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15, 7))
ax = np.hstack(ax)

i = 0
for key in images.keys():
    
    img = images[key][np.random.randint(0, len(images[key]))]
    ax[i].set_title(f'Image from {key}')
    ax[i].imshow(img)
    i += 1
    
plt.tight_layout()
plt.show()
```

- As we can see images are MRI slice-scans of human lungs

- Lets check amplitude distribution thoughout images 

```python
img = images['LUNG1-001'][0]

fig, ax = plt.subplots(1,2, figsize=(15, 5))
    
ax[0].imshow(img)
ax[0].set_title('Example of input image')

sns.distplot(img, ax=ax[1])
ax[1].set_title('Ampliturde distribution plot')
```

- Calculate some statistics

```python
stat = {}

for key in images.keys():
    
    imgs = images[key]
    
    stat[key] = (imgs.mean(), imgs.std(), imgs.min(), imgs.max())
    
mean_total = sum([values[0] for values in stat.values()])/len(stat.keys())
std_total = sum([values[1] for values in stat.values()])/len(stat.keys())
min_total = min([values[2] for values in stat.values()])
max_total = max([values[3] for values in stat.values()])
```

```python
# Statistics per folder
fig, ax = plt.subplots(2,5,figsize=(10, 5), sharey=True)
ax = np.hstack(ax)

i = 0
for key in images.keys():
    ax[i].bar(x=[f'mean\n{stat[key][0]:.0f}',
                 f'std\n{stat[key][1]:.0f}',
                 f'min\n{stat[key][2]:.0f}', 
                 f'max\n{stat[key][3]:.0f}'], 
              height=stat[key], 
              color=['red', 'green', 'blue', 'yellow'])
    
    ax[i].set_ylim(-1300, 4000)
    ax[i].set_yticks(np.arange(-1500, 4000, 1000))
    ax[i].set_title(f'{key}')
    i += 1
    
plt.tight_layout()
plt.show()
```

```python
# Total statistics
fig, ax = plt.subplots(1,1,figsize=(3, 3))

ax.bar(x=[f'mean\n{mean_total:.0f}',
          f'std\n{std_total:.0f}', 
          f'min\n{min_total:.0f}', 
          f'max\n{max_total:.0f}'], 
          height=[mean_total, std_total, min_total, max_total], 
          color=['red', 'green', 'blue', 'yellow'])
    
ax.set_ylim(-1300, 4000)
ax.set_yticks(np.arange(-1500, 4000, 1000))
ax.set_title(f'Average amplitude statistics')
plt.show()
```

- As we can see from plots above amplitude distribution along all data is pretty even with:

    - Mean equals -720
    - Standart Deviation 443
    - Maximum equals 3071
    - Minimum equals -1024
    

### Mask analysis

```python
masks = {}
total = 0

masks_files = {str(file) for file in 
               list(Path(os.path.join('../',config['dataset']['dir_masks'])).rglob("*.gz"))}


for file in sorted(masks_files):
    
    try:
        sub_folder = re.findall(r'LUNG[0-9]*-[0-9]*', file)[0]

        mask = nib.load(file)
        mask = mask.get_fdata().transpose(2, 0, 1)
        
        masks[sub_folder] = np.rot90(mask, axes=(1,2))

        print(f'{masks[sub_folder].shape[0]} files\tin {sub_folder} directory with shape {masks[sub_folder].shape[1]}x{masks[sub_folder].shape[2]}')
        total += masks[sub_folder].shape[0]
    except:
        print(f'ERROR: error reading {file} file')
    
print(f'\nTotal amount of masks: {total}')
```

```python
np.random.seed(333)

fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15, 7))
ax = np.hstack(ax)

i = 0
for key in masks.keys():
    
    img = masks[key][np.random.randint(0, len(masks[key]))]
    ax[i].imshow(img)
    i += 1
    
plt.tight_layout()
plt.show()
```

- Masks indicates pleural effusion 

- Lets check how mask corresponds to image

```python
img = images['LUNG1-026'][20]
mask = masks['LUNG1-026'][20]

plt.imshow(img)
plt.title('Scan without mask')
plt.show()

plt.imshow(mask*1000+img)
plt.title('Scan with mask rotated 90 degrees')
plt.show()
```

- Analyse how many masks have any values e.g. corresponds to images with pleural effusion

```python
mask_values = {key:[1 if mask.sum()>0 else 0 for mask in masks[key]] 
                            for key in masks.keys()}
```

```python
fig, ax = plt.subplots(5,2, figsize=(5,10), sharey=True)
ax = np.hstack(ax)

i = 0
for key in mask_values.keys():
    sns.countplot(x=mask_values[key], ax = ax[i])
    ax[i].set_title(f'{key}')
    ax[i].set_xlabel('Has effusion')
    i += 1
    
plt.tight_layout()
plt.show()
```

```python
total = []
mask_pos = []


for key in mask_values.keys():
    
    total.append(len(mask_values[key]))
    mask_pos.append(mask_values[key].count(1))
    
    print(f'In {key} folder percent of non-zero masks: {mask_pos[-1]/total[-1]:.0%}')

print(f'\nTotal percent of non-zero-masks: {sum(mask_pos)/sum(total):.0%}')
```

- Overal dataset has small inbalance towards samples with zero-value masks e.g. samples without pleural effusion
